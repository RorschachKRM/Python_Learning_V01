# 1. 两数之和

给定一个整数数组 nums 和一个整数目标值 target，请你在该数组中找出 和为目标值 target  的那 两个 整数，并返回它们的数组下标。
你可以假设每种输入只会对应一个答案，并且你不能使用两次相同的元素。
你可以按任意顺序返回答案。

示例 1：
输入：nums = [2,7,11,15], target = 9
输出：[0,1]
解释：因为 nums[0] + nums[1] == 9 ，返回 [0, 1] 。

示例 2：
输入：nums = [3,2,4], target = 6
输出：[1,2]

示例 3：
输入：nums = [3,3], target = 6
输出：[0,1]

````python
# 自解
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        List_len = len(nums)
        for i in range(List_len):
            for j in range(i + 1, List_len):
                if nums[i] + nums[j] == target:
                    return [i, j]
# O(n^2)  O(1)
````

````python
# 官方解法：哈希表
'''
时间复杂度较高的原因是寻找 target - x 的时间复杂度过高。因此，我们需要一种更优秀的方法，能够快速寻找数组中是否存在目标元素。如果存在，我们需要找出它的索引。
使用哈希表，可以将寻找 target - x 的时间复杂度降低到从 O(N) 降低到 O(1)。

这样我们创建一个哈希表，对于每一个 x，我们首先查询哈希表中是否存在 target - x，然后将 x 插入到哈希表中，即可保证不会让 x 和自己匹配。
'''

class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        hashtable = dict() # 创建一个空的哈希表
        for i, num in enumerate(nums): # 枚举 nums 中的每一个元素 num 和它的索引 i
            if target - num in hashtable: # 查询哈希表中是否存在 target - num
                return [hashtable[target - num], i] # 如果存在，返回 target - num 的索引和当前元素的索引
            hashtable[nums[i]] = i # 将当前元素 num 和它的索引 i 存入哈希表中
        return []
# O(n)  O(n)
````

## enumerate() 函数

- 作用：`enumerate()` 是 Python 内置函数，它将一个**可迭代对象**（如列表、元组、字符串）包装成**带索引的迭代器**，在遍历时同时获取 **索引** 和 **元素值**。

基本语法
```python
enumerate(iterable, start=0)
```
- `iterable`: 可迭代对象
- `start`: 索引起始值，默认为 0

本质：每次迭代返回一个 `(索引, 元素)` 的**元组**，通过 `for i, num in ...` 的元组解包语法，分别接收索引和值。


## 哈希表 / 散列表、哈希查找

### 1. 什么是哈希表？

哈希表是一种**键值对（key-value）**存储结构，它通过一个**哈希函数**把 key 转换成一个整数（哈希值），然后用这个整数直接定位到数据在内存中的存储位置。

**Python 中的 `dict`（字典）就是哈希表。**

### 2. 为什么需要哈希表？（以本题为例）

| 方案 | 查找 `target - num` 的方式 | 时间复杂度 |
|------|---------------------------|-----------|
| 暴力枚举 | 内层 for 循环遍历剩余数组 | O(n²) |
| **哈希表** | 用 key 直接查，O(1) | **O(n)** |

哈希表解决的问题：**如何快速判断一个元素是否存在，以及它在哪个位置？**
空间换时间

### 3. 哈希查找的原理

```
存储：key → 哈希函数 → 哈希值 → 内存中的桶(bucket)位置 → 存入
查找：key → 哈希函数 → 哈希值 → 同一位置 → 取出
```

查找是 **O(1)** 的。

#### 在 Python 中的体现
```python
hashtable = dict()

# 存储：key 是元素值，value 是索引
hashtable[2] = 0    # 存入 (2, 0)
hashtable[7] = 1    # 存入 (7, 1)

# 查找：直接用 key 判断是否存在
if 7 in hashtable:          # 哈希查找，O(1)
    return hashtable[7]     # 取出 value，O(1)
```

`7 in hashtable` 这个操作就是哈希查找——内部用哈希函数算出 `7` 对应的 bucket 位置，直接去那个位置看有没有数据，不需要遍历整个表。

### 4. 核心 Python 语法总结

| 语法 | 作用 | 示例 |
|------|------|------|
| `dict()` | 创建空字典 | `hashtable = dict()` |
| `key in dict` | 判断 key 是否存在（哈希查找，O(1)） | `if 7 in hashtable:` |
| `dict[key]` | 读取 key 对应的 value | `hashtable[7]` → `1` |
| `dict[key] = value` | 存入键值对 | `hashtable[2] = 0` |
| `enumerate(iterable)` | 迭代时同时获取索引和值 | `for i, num in enumerate(nums):` |

````python
'''包含哈希表、哈希函数、哈希查找的代码示例'''
class HashTable:
    """哈希表（链地址法实现）"""
    
    def __init__(self, size=10):
        """
        初始化哈希表
        :param size: 哈希表底层数组的大小（桶的数量）
        """
        self.size = size                # 桶的数量
        self.table = [[] for _ in range(size)]  # 每个桶是一个列表（链表）
    
    def hash_function(self, key):
        """
        哈希函数：除留余数法
        :param key: 要计算哈希值的键（要求键为整数，可自行扩展）
        :return: 哈希值（索引）
        """
        return key % self.size
    
    def put(self, key, value):
        """
        插入键值对（若键已存在，则更新值）
        :param key: 键
        :param value: 值
        """
        index = self.hash_function(key)
        bucket = self.table[index]
        
        # 检查键是否已存在，若存在则更新
        for i, (k, v) in enumerate(bucket):
            if k == key:
                bucket[i] = (key, value)
                return
        # 键不存在，追加新元素
        bucket.append((key, value))
    
    def get(self, key):
        """
        哈希查找：根据键查找对应的值
        :param key: 键
        :return: 找到则返回值，否则抛出 KeyError
        """
        index = self.hash_function(key)
        bucket = self.table[index]
        
        for k, v in bucket:
            if k == key:
                return v
        # 未找到
        raise KeyError(f"Key {key} not found")
    
    def __contains__(self, key):
        """支持 in 操作符"""
        try:
            self.get(key)
            return True
        except KeyError:
            return False
    
    def __setitem__(self, key, value):
        """支持 ht[key] = value 语法"""
        self.put(key, value)
    
    def __getitem__(self, key):
        """支持 ht[key] 语法"""
        return self.get(key)
    
    def __str__(self):
        """打印哈希表内容（用于调试）"""
        result = []
        for i, bucket in enumerate(self.table):
            if bucket:
                result.append(f"Bucket {i}: {bucket}")
        return "\n".join(result) if result else "Empty Hash Table"


# 演示与测试
if __name__ == "__main__":
    ht = HashTable(size=7)          # 创建哈希表，7个桶
    print("=== 插入键值对 ===")
    data = [(50, "A"), (700, "B"), (26, "C"), (33, "D"), (12, "E"), (19, "F")]
    for k, v in data:
        ht.put(k, v)
        print(f"插入 {k} -> {v} , 哈希值 = {ht.hash_function(k)}")
    
    print("\n=== 哈希表当前结构 ===")
    print(ht)
    
    print("\n=== 哈希查找演示 ===")
    test_keys = [33, 19, 100]
    for key in test_keys:
        try:
            value = ht.get(key)
            print(f"查找键 {key} 成功，对应值 = {value}")
        except KeyError:
            print(f"查找键 {key} 失败，该键不存在")
    
    print("\n=== 测试更新键值对 ===")
    ht.put(33, "X")      # 更新原key=33的值
    print(f"更新后，键 33 的值为：{ht[33]}")
    
    print("\n=== 使用 [] 语法操作 ===")
    ht[99] = "Z"
    print(f"ht[99] = {ht[99]}")
    print(f"66 in ht ? {66 in ht}")
    print(f"99 in ht ? {99 in ht}")
````
